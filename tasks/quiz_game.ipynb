```python
# Define the quiz questions and answers
quiz_questions = [
 {
        "question": "What is the capital of China?",
        "choices": ["Beijing", "Shanghai", "Hong Kong", "Guangzhou"],
        "correct_answer": "Beijing"
    },
    {
        "question": "What is the capital of India?",
        "choices": ["Mumbai", "Delhi", "Kolkata", "Bangalore"],
        "correct_answer": "Delhi"
    },
    {
        "question": "What is the capital of the United States?",
        "choices": ["New York", "Los Angeles", "Washington D.C.", "Chicago"],
        "correct_answer": "Washington D.C."
    },
    {
        "question": "What is the capital of Russia?",
        "choices": ["Moscow", "St. Petersburg", "Kazan", "Novosibirsk"],
        "correct_answer": "Moscow"
    },
    # Add more questions about capitals of famous countries
]

def play_quiz():
    # Initialize variables for tracking score and number of questions
    score = 0
    total_questions = len(quiz_questions)

    # Display welcome message and rules
    print("Welcome to the Quiz Game!")
    print("You'll be asked multiple-choice questions. Answer to the best of your knowledge.")

    # Iterate through each question in the quiz
    for question_data in quiz_questions:
        # Display the question and answer choices
        print(question_data["question"])
        for idx, choice in enumerate(question_data["choices"]):
            print(f"{idx + 1}. {choice}")

        # Get user's answer
        user_answer = input("Your answer (enter the option number): ")

        # Validate the user's answer
        if user_answer.isdigit() and 1 <= int(user_answer) <= len(question_data["choices"]):
            user_choice = question_data["choices"][int(user_answer) - 1]
            correct_answer = question_data["correct_answer"]

            # Evaluate the user's answer
            if user_choice == correct_answer:
                print("Correct!")
                score += 1
            else:
                print(f"Incorrect! The correct answer is: {correct_answer}")
        else:
            print("Invalid input. Skipping this question.")
        print()  # Add a line break

    # Display final results
    print("Quiz Finished!")
    print(f"Your score: {score}/{total_questions}")
    percentage = (score / total_questions) * 100
    if percentage >= 70:
        print("Great job! You did well.")
    else:
        print("Keep practicing. You can do better!")

    # Ask the user if they want to play again
    play_again = input("Do you want to play again? (yes/no): ")
    if play_again.lower() == "yes":
        play_quiz()  # Restart the game
    else:
        print("Thank you for playing!")

# Start the game
play_quiz()
```
